## Daily Volume for CME Energy Options Contracts for 11/02/2020
#### Below is a list of CME Energy Options contracts that have at least 100 lots of volume.
There are not many contracts that have significant volume


```python
import pandas as pd


df_cme_volume = pd.read_csv('cme_daily_volume_20201102.csv')
df_cme_volume['vol'] = pd.to_numeric(df_cme_volume.Volume.str.replace(',',''))
df_cme_volume['oi'] = pd.to_numeric(df_cme_volume['Open Interest'].str.replace(',',''))
df_cme_volume2 = df_cme_volume[(~df_cme_volume.Name.isna())]
tests = [
    df_cme_volume2.vol>10000,
    ~df_cme_volume2.Type.str.lower().str.contains('future'),
]
df_cme_volume2 = df_cme_volume2[[all(v) for v in zip(*tests)]]


df_cme_vol2 = df_cme_volume[
    (df_cme_volume.Type=='Options') & 
    (df_cme_volume.vol>=100)
].sort_values('vol',ascending=False)

pd.set_option('display.max_colwidth', 0)
df_cme_vol3 = df_cme_vol2[[c for c in df_cme_vol2 if c not in ['oi','vol']]]
df_cme_vol3
```
