### Download intra data from ICE

```python
import pandas as pd
import numpy as np
import zipfile
import requests
import io
import base64

d = '2020-03-11'
u = f'https://d1tll0q2bcilyo.cloudfront.net/CME%20EMINI/{d}/WATCHLIST_684_20200311.txt.bz2'
```

```python
import bz2
def get_from_ice(u):
    f = requests.get(u)
    data = bz2.decompress(f.content)
    lines = data.decode('UTF-8').split('\n')
    return lines
```

```python
nymex_url = 'https://d1tll0q2bcilyo.cloudfront.net/NYMEX/2020-03-25/PREMREF_686_20200325.txt.bz2'
all_nymex_lines = get_from_ice(nymex_url)
```
